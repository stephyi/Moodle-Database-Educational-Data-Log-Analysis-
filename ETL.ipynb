```python
##Using SQLalchemy for connection

def get_engine(db, user, host, passwd):
    """
    Get SQLalchemy engine using credentials.
    Input:
    db: database name
    user: Username
    host: Hostname of the database server
    port: Port number
    passwd: Password for the database
    """

    url = 'postgresql://{user}:{passwd}@{host}/{db}'.format(
        user=user, passwd=passwd, host=host, db=db)
    engine = create_engine(url, pool_size = 50)
    return engine
```

### Number of quiz submissions by hour of day

```python
%sql select count(id) as submission_count, EXTRACT(HOUR FROM to_timestamp(timecreated))from mdl_logstore_standard_log where action='submitted' AND component='mod_quiz' group by EXTRACT(HOUR FROM to_timestamp(timecreated));
```

### Monthly usage time of learners who have confirmed and are not deleted

```python
%%sql
SELECT DISTINCT(confirmed,deleted)
FROM mdl_user
LIMIT 5;
```

```python
%%sql
SELECT count(lastaccess - firstaccess) AS MonthlyUsageTime, EXTRACT (MONTH FROM to_timestamp(firstaccess)) AS Month
from mdl_user
where confirmed=1 and deleted=0
GROUP BY month;
```

### Python class to:
* Overall grade of learners for
* Number of forum posts

```python
class get_info_():
    
    
    def __init__(self,host,port,db,user,password):
        self.host=host
        self.port=port
        self.db= db
        self.user = user
        self.password=password
        
        self.create_connection()
        
    def create_connection(self):
    ## Using psycopg2 for connection
     self.conn = psycopg2.connect(host=self.host, port = self.port, database=self.db ,user=self.user,password=self.password)
    

    # function to get average grade
    def get_average_grade(self):
        avg_grade = pd.read_sql_query( sql = "SELECT AVG(finalgrade) AS avg_grade FROM mdl_grade_grades", 
                         con = self.conn)
        return avg_grade
    
    # function to get number of forum posts
    def get_forum_posts(self):
        forum_posts = pd.read_sql_query( sql = "SELECT COUNT(id) AS forum_post_counts  FROM mdl_forum_posts;", 
                         con = self.conn)
        return forum_posts
        
    conn.close()
```

### Dedication Time
* The time between login and last activity for that login session, which is the last activity before the next login event for a given learner.  
* Moodle estimates time based in the concepts of Session and Session duration applied to Moodle's log entries:

  * Click: every time that a user accesses a page in Moodle a log entry is stored.
  * Session: set of two or more consecutive clicks in which the elapsed time between every pair of consecutive clicks does not overcome an established maximum time.
  * Session duration: elapsed time between the first and the last click of the session.

### 2.Login and activity counts

```python

%%sql
SELECT id,COUNT(action) as loggedInCount
FROM mdl_logstore_standard_log
WHERE action='loggedin'
GROUP BY id
ORDER BY loggedInCount DESC
LIMIT 10;
```
